**Esse notebook tem como objetivo organizar os dados de saída dos CPs do arquivo CProp.txt gerado pelo software Mwfn.**

```python
import pandas as pd #importando a biblioteca Pandas
import re  #importando a biblioteca Regex
import warnings #importando a biblioteca warnings

warnings.filterwarnings('ignore') ## ocultando os warnings que aparecem em algumas células
```

```python
## Atribuindo os parâmetros do arquivo 'CPprop.txt', que serão requisitados utilizando a função 'compile 'o Regex nas variáveis abaixo

t = re.compile(r'\sLaplacian of electron density:\s')  #atribuindo à string "Laplacian of electron density:" na variável 't'

r = re.compile(r'\sCP\s')  #atribuindo à string "CP e Pair" na variável 'r'

a = re.compile(r'\sEigenvalues of Hessian:\s')  #atribuindo à string "Eigenvalues of Hessian:" na variável 'a'

p = re.compile(r'\sPosition\s')  #atribuindo à string "Position:" na variável 'p'

#criando listas vazias para serem preenchidas de acordo com a quantidade de variáveis utilizadas acima
parametros_cp=[]
autovalores_cp=[]
laplaciano_cp=[]
coordenadas_cp=[]

with open('CPprop.txt', 'r') as arquivo: #abrindo o 'CPprop.txt' com a varivável 'arquivo'

    for linha in arquivo: # varrendo as linhas do 'arquivo' com a variável 'linha'

        if r.search(linha):  # se na varivel 'linha' encontrar a string 'CP e Pair'  então
            # armazenando as informações do CP na variável 'cp' e limpando caracteres desnecessários com replace
            cp=linha.replace('CP', "").replace(" ", "").replace('--', '').replace('\n', '').replace('Type', '')

            # preenchendo a lista 'parametros_cp' com as informações da variável 'cp' com a função 'append'
            parametros_cp.append((cp))

        if a.search(linha): # se na variável 'linha' encontrar a strig 'Eigenvalues of Hessian:' então
             #armazenando e limpando as informações na varivel 'autovalor' e subdividindo as informações em sublistas com a função split
             autovalor=linha.replace('Eigenvalues of Hessian:','').split()

             autovalores_cp.append(autovalor) ## preenchendo a lista com as informações da variável autovalor

        if t.search(linha): # se na variável 'linha' encontrar a strig 'Laplacian of electron density:' então
             #armazenando e limpando as informações na varivel 'laplacian'
            laplacian=linha.replace('Laplacian of electron density:', '').replace(" ","").replace('\n','')

            ## preenchendo a lista 'laplaciano_cp' com as informações da variável 'laplacian' com a função 'append'
            laplaciano_cp.append((laplacian))

        if p.search(linha): # se na variável 'linha' encontrar a strig 'coordenadas_cp:' então
            #armazenando e limpando as informações na varivel 'position'
            position=linha.replace('Position (Bohr):','').replace('\n','').strip().split()

            ## preenchendo a lista 'coordenadas_cp' com as informações da variável 'position' com a função 'append'
            coordenadas_cp.append(position)

#printando as listas preenchidas apenas para verificar se todas a listas foram preenchidas corretamente com os seus respectivos parâmetros

#As listas preenchidas estão comentadas apenas para não ficar com  muitas informações na tela

#parametros_cp
#autovalores_cp
#laplaciano_cp
#coordenadas_cp
```

```python
## criando um df 'df_coord_bohr' a partir da lista preenchida 'coordenadas_cp' que conterá as informações das coordenadas em bohr dos CPs
df_coord_bohr = pd.DataFrame(coordenadas_cp, columns=['(X)', '(Y)', '(Z)'])
df_coord_bohr #printando o df 'df_coord_bohr'
```

```python
## criando um df 'df_auto_valor' a partir da lista preenchida 'autovalores_cp'  que conterá as informações dos autovalores de cada CP
df_auto_valor = pd.DataFrame(autovalores_cp, columns=['Lambda 1', 'Lambda 2', 'Lambda 3'])
df_auto_valor
```

```python
#criando um df 'df_laplacian' a partir da lista preenchida 'laplaciano_cp' que conterá as informações do laplaciano de cada CP
df_laplacian = pd.DataFrame(laplaciano_cp, columns=['Laplacian of electron density'])
df_laplacian
```

```python
## criando um df  a partir da lista preenchida 'parametros_cp'  que conterá as informações desorganizadas do label do 'CP', 'Pair', 'Rank' e 'Signature'
df = pd.DataFrame(parametros_cp, columns = ['CP'])
df
```

```python
## Separando o df utilizando o comando split por meio do caracter ',' e gerando a coluna 'Pair' da coluna 'CP' do df anterior
df[['CP', 'Pair', '']]=df['CP'].str.split(',', expand=True)
df
```

```python
## Concatenando as colunas 'Pair' e a coluna '' por meio do ';' para organizar a coluna 'Pair' do df
df["Pair"] = df["Pair"] + ";" + df[""]
df
```

```python
## deletando a última coluna [''] (não tem utilidade essa coluna)
df = df.drop('', axis=1)
df
```

```python
## gerando a coluna 'Rank' e  a 'Signature' a partir da coluna 'Pair' do df
df[['Rank', 'Signature']]=df['Pair'].str.replace('(', '').str.replace(')', '').str.split(';', expand=True)
df
```

```python
df ## printando df
```

```python
## criando um dicionário 'type_cp'
type_cp={'(3;-3)': 'Nuclear',
         '(3;-1)': 'Ligação',
         '(3;+3)': 'Cela',
         '(3;+1)': 'Anel'}
## criando uma coluna ''Type of CP' a partir da coluna 'Pair' do df utilizando o dicionário 'tipo_cp'
df['Type of CP']=df['Pair'].map(type_cp)
```

```python
df=df[['CP','Rank', 'Signature' ,'Pair', 'Type of CP']] ## reorganizando as colunas do df

df
```

```python
## Concatenando os DataFrame 'df_auto_valor', 'df', 'df_laplacian, df_coord_bohr'
## passando como flag axis=1 (colunas)
df = pd.concat([df_auto_valor, df, df_laplacian, df_coord_bohr], axis=1)

## Reorganizando todas as colunas do df
df=df[['CP', '(X)', '(Y)', '(Z)', 'Rank', 'Signature', 'Pair', 'Type of CP', 'Lambda 1','Lambda 2','Lambda 3', 'Laplacian of electron density']]


```

```python
df ##printando o df com as colunas organizadas
```

```python
## criando uma lista 'parametros_float' com as colunas do df que serão convertidas em float
parametros_float=['Lambda 1', 'Lambda 2', 'Lambda 3', 'Laplacian of electron density','(X)', '(Y)', '(Z)']

## Obs: Não precisava de ter criado essa lista e esse laço que está abaixo,
## isso foi feito apenas para não ficar uma linha de código muito grande dentro do laço

for troca_float in parametros_float: ## varrendo a lista 'parametros_float'
  df[[troca_float]] = df[[troca_float]].astype(float) ## fazendo um casting em cada elemento da lista


## transformando as colunas 'CP', 'Rank' e 'Signature' em int
df[['CP','Rank','Signature']] = df[['CP','Rank','Signature']].astype(int)
```
